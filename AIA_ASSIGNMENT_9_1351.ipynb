```python
# Building a Chat-bot using IBM Watson visual studio
from ibm_watson import AssistantV2
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator


authenticator = IAMAuthenticator('your_api_key')
assistant = AssistantV2(
    version='your_version',
    authenticator=authenticator
)

assistant.set_service_url('your_service_url')

session = assistant.create_session('your_assistant_id').get_result()

print("Welcome to the chatbot! How can I assist you today?")

while True:
    user_input = input()

    response = assistant.message(
        assistant_id='your_assistant_id',
        session_id=session['session_id'],
        input={
            'message_type': 'text',
            'text': user_input
        }
    ).get_result()

    assistant_response = response['output']['generic'][0]['text']

    print(assistant_response)

    if user_input.lower() == 'exit':
        break

assistant.delete_session(
    assistant_id='your_assistant_id',
    session_id=session['session_id']
)
```

```python
#Building a Chat-bot using Pandora bots
import requests

botname = 'your_botname'
app_id = 'your_app_id'
user_key = 'your_user_key'
input_text = ''

print("Welcome to the chatbot! How can I assist you today?")

while True:
    input_text = input()

    response = requests.get(
        'https://aiaas.pandorabots.com/talk/%s?input=%s&user_key=%s&app_id=%s' %
        (botname, input_text, user_key, app_id)
    )

    bot_response = response.json()[0]['response']

    print(bot_response)

    if input_text.lower() == 'exit':
        break
```

```python
#Build a virtual assistant for Wikipedia using Wolfram Alpha and Python
import wolframalpha
import wikipedia

app_id = 'your_app_id'

print("Welcome to the virtual assistant! How can I assist you today?")

while True:
    # Get user input
    user_input = input()

    client = wolframalpha.Client(app_id)
    res = client.query(user_input)

    if len(res.pods) > 0:
        for pod in res.pods:
            if pod.text:
                print(pod.text)
                break
    else:
        try:
            wikipedia_result = wikipedia.summary(user_input, sentences=1)
            print(wikipedia_result)
        except wikipedia.exceptions.PageError:
            print("Sorry, I couldn't find anything on that topic.")

    if user_input.lower() == 'exit':
        break
```
